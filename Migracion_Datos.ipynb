```python
import pandas as pd
from pymongo import MongoClient

# 1. Conectarse a MongoDB Atlas
client_atlas = MongoClient("mongodb+srv://edisonescobar01:1234@cluster0.uzpaxxp.mongodb.net/")
db_atlas = client_atlas["Conciertos"]

# 2. Leer y limpiar las colecciones
df_musica_conciertos = pd.DataFrame(list(db_atlas["musica_conciertos"].find()))
df_concierto = pd.DataFrame(list(db_atlas["Concierto"].find()))

# 3. Eliminar columna '_id' para evitar conflictos al insertar
df_musica_conciertos.drop(columns=['_id'], inplace=True, errors='ignore')
df_concierto.drop(columns=['_id'], inplace=True, errors='ignore')

# 4. Opcional: filtrar columnas específicas (para evitar mezcla)
columnas_musica = ['track', 'artist', 'seeds', 'number_of_emotion_tags', 'valence_tags', 'arousal_tags', 'dominance_tags']
columnas_concierto = ['Usuario', 'Fecha_Creación', 'Comentario', 'Polaridad', 'Sentimiento']

# Aplicar filtros
df_musica_conciertos = df_musica_conciertos[[col for col in columnas_musica if col in df_musica_conciertos.columns]]
df_concierto = df_concierto[[col for col in columnas_concierto if col in df_concierto.columns]]

# 5. Conectarse a MongoDB local
client_local = MongoClient("mongodb://localhost:27017/")
db_local = client_local["base_unificada"]

# 6. Insertar en colecciones separadas
# Borrar contenido anterior (opcional)
db_local["musica_conciertos"].delete_many({})
db_local["Concierto"].delete_many({})

# Insertar nuevos datos
db_local["musica_conciertos"].insert_many(df_musica_conciertos.to_dict('records'))
db_local["Concierto"].insert_many(df_concierto.to_dict('records'))

print("✅ Migración exitosa: datos copiados de MongoDB Atlas a MongoDB local.")
```

```python
import pandas as pd
from pymongo import MongoClient
from sqlalchemy import create_engine
import urllib

# 1. Conectar a MongoDB
cliente_mongo = MongoClient("mongodb://localhost:27017/")
bd_mongo = cliente_mongo["base_unificada"]  

# 2. Conexión a SQL Server Express con autenticación de Windows
params = urllib.parse.quote_plus(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=localhost\\sqlexpress;"
    "DATABASE=Base_Proyecto;"  
    "Trusted_Connection=yes;"
)

motor_sql = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

# 3. Lista de colecciones a migrar
colecciones = ["comentarios_anime2"]  # <-- cambia por las reales

# 4. Migrar cada colección
for nombre_col in colecciones:
    datos = list(bd_mongo[nombre_col].find())

    if datos:
        df = pd.DataFrame(datos)
        df.drop(columns=["_id"], inplace=True, errors="ignore")
        df.to_sql(nombre_col, motor_sql, if_exists="replace", index=False)
        print(f"✅ Migración de '{nombre_col}' completada.")
    else:
        print(f"⚠️ La colección '{nombre_col}' está vacía.")
```
